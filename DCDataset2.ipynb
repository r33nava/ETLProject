```python
#Datasource URL: https://www.neighborhoodinfodc.org/comparisontables/comparisontables.html

# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import requests
import json
from pprint import pprint
```

```python
#Source #2: https://www.neighborhoodinfodc.org/comparisontables/comparisontables.html
#Ward Well-Being
#Load Data

wardwb_df = pd.read_csv("WardWellBeing.csv")
wardwb_df.head(8)
```

```python
# Collecting a list of all columns within the DataFrame
wardwb_df.columns
```

```python
# Reorganizing the columns using double brackets

wardwb_data = wardwb_df[["WARD2012","PctPoorPersons_2011_15","PctUnemployed_2011_15","AvgFamilyIncAdj_2011_15"]]
wardwb_data.head(10)
```

```python
# Using .rename(columns={}) in order to rename columns
renamed_df = wardwb_data.rename(columns={"WARD2012":"WARD", "PctPoorPersons_2011_15":"PercentPoor","PctUnemployed_2011_15":"PercentUnemployed","AvgFamilyIncAdj_2011_15":"AvgFamilyIncome"})
renamed_df.head(9)
```

```python
wardwb_data = renamed_df.dropna()
```

```python
wardwb_data
```

```python
#Source #2: https://www.neighborhoodinfodc.org/comparisontables/comparisontables.html
#Precinct Well Being
#Load Data

precinctwb_df = pd.read_csv("PrecintWellBeing.csv")
precinctwb_df.head()
```

```python
# Collecting a list of all columns within the DataFrame
precinctwb_df.columns
```

```python
# Reorganizing the columns using double brackets

precinctwb_data = precinctwb_df[["VOTERPRE2012","PctPoorPersons_2011_15","PctUnemployed_2011_15","AvgFamilyIncAdj_2011_15"]]
precinctwb_data.head()
```

```python
# Using .rename(columns={}) in order to rename columns
precinctwb_data = precintwb_data.rename(columns={"VOTERPRE2012":"VOTING_PRECINCT", "PctPoorPersons_2011_15":"PercentPoor","PctUnemployed_2011_15":"PercentUnemployed","AvgFamilyIncAdj_2011_15":"AvgFamilyIncome"})
```

```python
precinctwb_data.head()
```

```python
# Export files as a CSV, without the Pandas index, but with the header
wardwb_data.to_csv("DCWards_cleaned.csv", index=False, header=True)
precinctwb_data.to_csv("DCPrecincts_cleaned.csv", index=False, header=True)
```
